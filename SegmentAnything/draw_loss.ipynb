```python
import re
import matplotlib.pyplot as plt


def extract_train_loss(file_path):
    train_loss_list = []
    
    with open(file_path, 'r') as file:
        for line in file:
            match = re.search(r'Train Loss: (-?\d+\.\d+)', line)
            if match:
                train_loss = float(match.group(1))
                train_loss_list.append(train_loss)
    
    plt.plot(train_loss_list)
    plt.xlabel('Epoch')
    plt.ylabel('Train Loss')
    plt.title('Train Loss Curve')
    plt.show()
    
    return train_loss_list


file_path = 'logs/train_logs/MySam/MySam/BinaryMaskLoss/lr0.001_bs1_pwTrue_fNothing_add/epoch/2024-04-14-01.18/log.txt'
train_loss_list = extract_train_loss(file_path)
print(train_loss_list)
```
